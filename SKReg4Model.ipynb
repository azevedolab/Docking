# **SKReg4Model: Scikit-Learn Regressors for Modeling**

This Jupyter Notebook builds multiple regression models using methods based on a combination of features calculated for datasets presented as CSV (comma-separated values) files ([de Azevedo et al., 2024](https://doi.org/10.1002/jcc.27449); [Géron, 2023](https://www.isbns.net/isbn/9781098125974/)). To generate the regression models, it employs Regressor classes from the [Scikit-Learn](https://scikit-learn.org/stable/) library ([Pedregosa et al., 2011](https://doi.org/10.48550/arXiv.1201.0490)). Each regression model generated by this code resulted from a combination of features to sweep the scoring function space ([Ross et al., 2013](https://doi.org/10.1021/ct4004228); [Bitencourt-Ferreira et al., 2024](https://doi.org/10.2174/0929867330666230321103731)). This approach allows an intelligent scanning of large regions of the scoring function space, which increases the chances of finding an adequate model for a targeted protein system ([Bitencourt-Ferreira & de Azevedo, 2019](https://doi.org/10.1007/978-1-4939-9752-7_17)). It employs the metrics recommended by [Walsh et al., 2021](https://doi.org/10.1038/s41592-021-01205-4) and [de Azevedo et al., 2024](https://doi.org/10.1002/jcc.27449) to evaluate the model's predictive performance. You may also build a single regression model using a specific dataset dictionary (e.g., cdk2ki_single), which comprises a fixed set of features. To build regression models, this Jupyter Notebook reads features (energy terms, pair interactions, and descriptors) from a CSV file (e.g., [CDK2 with K<sub>i</sub> (Training Set)](https://bit.ly/CDK2_Ki_Training_Set)) generated with Molegro Virtual Docker ([Thomsen & Christensen, 2006](https://doi.org/10.1021/jm051197e)), AutoDock4 ([Morris et al.,2009](https://doi.org/10.1002/jcc.21256)), Vinardo ([Quiroga & Villarreal, 2016](https://doi.org/10.1371/journal.pone.0155183)), AutoDockVina 1.2 ([Eberhardt et al., 2021](https://doi.org/10.1021/acs.jcim.1c00203)), and SAnDReS 2.0 ([de Azevedo et al., 2024](https://doi.org/10.1002/jcc.27449)).

**Define cross_validation() function**

**cross_validation(model,X,y,n_splits,random_state,verbose)**
<br> </br>

Function to set up a K-fold class from the [Scikit-Learn](https://scikit-learn.org/stable/) library ([Pedregosa et al., 2011](https://doi.org/10.48550/arXiv.1201.0490); [Coelho et al., 2018](https://www.isbns.net/isbn/9781788623223/)). This K-fold class builds an n-fold cross-validation loop and tests the generalization ability of regression. We generally obtain a more conservative estimate (that is, we have poorer predictive metrics) with cross-validation. The cross-validation estimate is a better model for generalizing and predicting unseen data.



    Parameters:

    model (scikit-learn model):     Regression model generated with Scikit-Learn
                                    library ([Pedregosa et al., 2011]
                                    (https://doi.org/10.48550/arXiv.1201.0490))
    X (dataframe):                  X data
    y (dataframe):                  y data
    n_splits (integer):             Number of splits
    random_state (integer):         Random seed
    verbose(boolean):               Generate additional information about
                                    cross-validation() funcitons

<br> </br>

<br> </br>

```python
def cross_validation(model,X,y,n_splits,random_state,verbose):
    """Function to set up a K-fold class from the Scikit-Learn library
    ([Pedregosa et al., 2011](https://doi.org/10.48550/arXiv.1201.0490).
    """
    # Import section
    from sklearn.model_selection import KFold, cross_val_score
    import numpy as np

    # Set up k-fold class
    kf = KFold(n_splits=n_splits,shuffle=True, random_state=random_state)

    # Looping through kf.split()
    for train,test in kf.split(X):

        # Generate regression model
        model.fit(X[train],y[train])

    # Show Walsh metrics if requested
    if verbose:
        # Show average coefficient of determination using n-fold crossvalidation
        scores = cross_val_score(model,X,y,cv=kf)
        msg_out = "Average coefficient of determination using n-fold "
        msg_out += "cross-validation"
        print("\n"+msg_out+":",np.mean(scores))

    # Return model
    return model
```

```python
def Walsh_and_de_Azevedo_regression_metrics(y,y_pred):
    """Function to calculate regression metrics as recommended by
    [Walsh et 2021](https://pubmed.ncbi.nlm.nih.gov/34316068/)
    and
    [de Azevedo Jr et al., 2024](https://pubmed.ncbi.nlm.nih.gov/38900052/)).
    """
    # Import section
    from sklearn.metrics import median_absolute_error
    from scipy import stats
    import numpy as np

    # Determine metrics
    n = len(y)
    y_bar = y.mean()
    rss = np.sum( np.square(y_pred - y) )
    rmse =  np.sqrt( (np.sum( (y - y_pred)**2 ))/n )
    ss_res = np.sum( (y - y_pred)**2 )
    ss_tot = np.sum( (y - y_bar)**2 )
    R2 = 1 - ss_res/ss_tot
    mae = median_absolute_error(y,y_pred)
    r_pearson,p_pearson = stats.pearsonr(y,y_pred)
    rho,p_spearman = stats.spearmanr(y,y_pred)
    core_dome = rmse**2 + mae**2 + (R2 -1)**2
    dome =    np.sqrt( core_dome )
    edomer2 = np.sqrt( core_dome + (r_pearson**2 - 1)**2 )
    edomerho =np.sqrt( core_dome + (rho**2 - 1)**2 )
    edome =   np.sqrt( core_dome + (r_pearson**2 - 1)**2 + (rho**2 - 1)**2 )

    # Return metrics
    return r_pearson,p_pearson,rho,p_spearman,rss,rmse,mae,R2,dome,edomer2,\
    edomerho,edome
```

```python
def scatter_plot(y,y_pred,
                n_points,point_type,
                color4data,color4models,
                xlabel,ylabel,title,
                xlabel_fontsize,ylabel_fontsize,title_fontsize,
                xticks_fontsize,yticks_fontsize,
                xlim,ylim,
                grid_on,
                plt_out,plt_dpi):
    """Function to generate a basic scatter plot with data points and a
    regression model line. It relies on [matplotlib](https://matplotlib.org/)
    library to build a requested plot.
    """
    # Import section
    import numpy as np
    import matplotlib.pyplot as plt

    # Obtain m (slope) and b(intercept) of a regression line
    m, b = np.polyfit(y,y_pred, 1)

    # Set up an array
    y_in = np.linspace(xlim[0], xlim[1], num = n_points)

    # Create basic scatterplot
    plt.plot(y, y_pred,point_type,color = color4data)

    # Use color4models as color for regression model
    plt.plot(y_in, m*y_in+b, color=color4models)
    plt.xlabel(xlabel,fontsize = xlabel_fontsize)
    plt.ylabel(ylabel,fontsize = ylabel_fontsize)
    plt.title(title,fontsize = title_fontsize)
    plt.xticks(fontsize = xticks_fontsize)
    plt.yticks(fontsize = yticks_fontsize)
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.grid(grid_on)
    plt.savefig(plt_out,dpi = plt_dpi)
    plt.show()
    plt.close()
```

```python
class RegressorModels(object):
    """Class to build regression models using a combination of features and the
    Regressor classes from the [Scikit-Learn](https://scikit-learn.org/stable/)
    library
    ([Pedregosa et al., 2011](https://doi.org/10.48550/arXiv.1201.0490)).
    """
    # Define constructor method
    def __init__(self,
        r_sk,
        training_set,
        test_set,
        y_train,
        y_test,
        features,
        l,
        header,
        t_root,
        param_dict,
        miscellaneous,
        dataset):
        """Constructor method"""

        # Define attributes (miscellaneous[""])
        # For regression method
        self.r_sk = r_sk

        # For dataset
        self.training_set = training_set
        self.test_set = test_set
        self.y_train = y_train
        self.y_test = y_test
        self.features = features
        self.l = l
        self.header = header
        self.t_root = t_root
        self.param_dict = param_dict
        self.miscellaneous = miscellaneous
        self.dataset = dataset

        # For cross validation
        self.random_state_cv = self.miscellaneous["random_state_cv"]
        self.n_splits_cv = self.dataset["n_splits"]

        # For scatter plot
        self.n_points = self.miscellaneous["n_points"]
        self.point_type = self.miscellaneous["point_type"]
        self.color4data = self.miscellaneous["color4data"]
        self.color4models = self.miscellaneous["color4models"]
        self.scatter_xlabel = self.miscellaneous["scatter_xlabel"]
        self.scatter_ylabel = self.miscellaneous["scatter_ylabel"]
        self.xlabel_fontsize = self.miscellaneous["xlabel_fontsize"]
        self.ylabel_fontsize = self.miscellaneous["ylabel_fontsize"]
        self.title_fontsize = self.miscellaneous["title_fontsize"]
        self.xticks_fontsize = self.miscellaneous["xticks_fontsize"]
        self.yticks_fontsize = self.miscellaneous["yticks_fontsize"]
        self.xlim = self.miscellaneous["xlim"]
        self.ylim = self.miscellaneous["ylim"]
        self.grid_on = self.miscellaneous["grid_on"]
        self.plt_dpi = self.miscellaneous["plt_dpi"]

    # Define show() method
    def show(self):
        """Method to show all sklearn regressor classes available in this
        code."""
         # Import section
        from sklearn.utils import all_estimators
        import sklearn

        # Get regressors
        complete_list = all_estimators(type_filter='regressor')

        # Show message
        msg_out = "\nList of regressors available in Scikit-Learn ("
        msg_out +=  str(sklearn.__version__)+") implemented in this code:"
        print(msg_out)
        # Looping through estimators
        i = 1
        for name, class_ in complete_list:
            if "CV" not in str(class_.__name__) and \
                "Multi" not in str(class_.__name__) and \
                "RegressorChain" not in str(class_.__name__) and \
                "IsotonicRegression" not in str(class_.__name__) and \
                "RadiusNeighborsRegressor" not in str(class_.__name__) and \
                "DummyRegressor" not in str(class_.__name__):
                print(f'{i}. {class_.__name__}')
                i += 1

    # Define pick_Regressor() method
    def pick_Regressor(self,X_train):
        """Method to select a Regressor class from the Scikit-Learn library
        ([Pedregosa et al., 2011](https://doi.org/10.48550/arXiv.1201.0490)).
        """
        # Import section
        import numpy as np
        from sklearn.ensemble._weight_boosting import AdaBoostRegressor
        from sklearn.linear_model._bayes import ARDRegression
        from sklearn.ensemble._bagging import BaggingRegressor
        from sklearn.linear_model._bayes import BayesianRidge
        from sklearn.cross_decomposition._pls import CCA
        from sklearn.tree._classes import DecisionTreeRegressor
        from sklearn.linear_model._coordinate_descent import ElasticNet
        from sklearn.tree._classes import ExtraTreeRegressor
        from sklearn.ensemble._forest import ExtraTreesRegressor
        from sklearn.linear_model._glm.glm import GammaRegressor
        from sklearn.gaussian_process._gpr import GaussianProcessRegressor
        from sklearn.ensemble._gb import GradientBoostingRegressor
        from sklearn.ensemble._hist_gradient_boosting.gradient_boosting import \
                                                HistGradientBoostingRegressor
        from sklearn.linear_model._huber import HuberRegressor
        from sklearn.neighbors._regression import KNeighborsRegressor
        from sklearn.kernel_ridge import KernelRidge
        from sklearn.linear_model._least_angle import Lars
        from sklearn.linear_model._coordinate_descent import Lasso
        from sklearn.linear_model._least_angle import LassoLars
        from sklearn.linear_model._least_angle import LassoLarsIC
        from sklearn.linear_model._base import LinearRegression
        from sklearn.svm._classes import LinearSVR
        from sklearn.neural_network._multilayer_perceptron import MLPRegressor
        from sklearn.svm._classes import NuSVR
        from sklearn.linear_model._omp import OrthogonalMatchingPursuit
        from sklearn.linear_model._passive_aggressive import \
                                                    PassiveAggressiveRegressor
        from sklearn.cross_decomposition._pls import PLSCanonical
        from sklearn.cross_decomposition._pls import PLSRegression
        from sklearn.linear_model._glm.glm import PoissonRegressor
        from sklearn.linear_model._quantile import QuantileRegressor
        from sklearn.ensemble._forest import RandomForestRegressor
        from sklearn.linear_model._ransac import RANSACRegressor
        from sklearn.linear_model._ridge import Ridge
        from sklearn.linear_model._ridge import RidgeCV
        from sklearn.linear_model._stochastic_gradient import SGDRegressor
        from sklearn.svm._classes import SVR
        from sklearn.ensemble._stacking import StackingRegressor
        from sklearn.linear_model._theil_sen import TheilSenRegressor
        from sklearn.compose._target import TransformedTargetRegressor
        from sklearn.linear_model._glm.glm import TweedieRegressor
        from sklearn.ensemble._voting import VotingRegressor

        # Select a Regressor
        if self.r_sk == "adaboostregressor" or \
                        self.r_sk == "adaboostregressorcv":
            # Instantiate an object of AdaBoostRegressor() class
            self.sklearn_model = AdaBoostRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "ardregression" or \
                        self.r_sk == "ardregressioncv":
            # Instantiate an object of ARDRegression() class
            self.sklearn_model = ARDRegression(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "baggingregressor" or \
                        self.r_sk == "baggingregressorcv":
            # Instantiate an object of BaggingRegressor() class
            self.sklearn_model = BaggingRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "bayesianridge" or \
                        self.r_sk == "bayesianridgecv":
            # Instantiate an object of BayesianRidge() class
            self.sklearn_model = BayesianRidge(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "cca" or self.r_sk == "ccacv":
            # Instantiate an object of CCA() class
            self.sklearn_model = CCA(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "decisiontreeregressor" or \
                        self.r_sk == "decisiontreeregressorcv":
            # Instantiate an object of DecisionTreeRegressor() class
            self.sklearn_model = DecisionTreeRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "elasticnet" or \
                        self.r_sk == "elasticnetcv":
            # Instantiate an object of ElasticNet() class
            self.sklearn_model = ElasticNet(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "extratreeregressor" or \
                        self.r_sk == "extratreeregressorcv":
            # Instantiate an object of ExtraTreeRegressor() class
            self.sklearn_model = ExtraTreeRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "extratreesregressor" or \
                        self.r_sk == "extratreesregressorcv":
            # Instantiate an object of ExtraTreesRegressor() class
            self.sklearn_model = ExtraTreesRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "gammaregressor" or \
                        self.r_sk == "gammaregressorcv":
            # Instantiate an object of GammaRegressor() class
            self.sklearn_model = GammaRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "gaussianprocessregressor" or \
                        self.r_sk == "gaussianprocessregressorcv":
            # Instantiate an object of GaussianProcessRegressor() class
            self.sklearn_model = GaussianProcessRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "gradientboostingregressor" or \
                        self.r_sk == "gradientboostingregressorcv":
            # Instantiate an object of GradientBoostingRegressor() class
            self.sklearn_model = GradientBoostingRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "histgradientboostingregressor" or \
                        self.r_sk == "histgradientboostingregressorcv":
            # Instantiate an object of HistGradientBoostingRegressor() class
            self.sklearn_model = HistGradientBoostingRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "huberregressor" or \
                        self.r_sk == "huberregressorcv":
            # Instantiate an object of HuberRegressor() class
            self.sklearn_model = HuberRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "kneighborsregressor" or \
            self.r_sk == "kneighborsregressorcv":
            # Instantiate an object of KNeighborsRegressor() class
            self.sklearn_model = KNeighborsRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "kernelridge" or \
            self.r_sk == "kernelridgecv":
            # Instantiate an object of KernelRidge() class
            self.sklearn_model = KernelRidge(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "lars" or self.r_sk == "larscv":
            # Instantiate an object of Lars() class
            self.sklearn_model = Lars(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "lasso" or self.r_sk == "lassocv":
            # Instantiate an object of Lasso() class
            self.sklearn_model = Lasso(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "lassolars" or \
            self.r_sk == "lassolarscv":
            # Instantiate an object of LassoLars() class
            self.sklearn_model = LassoLars(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "lassolarsic" or \
            self.r_sk == "lassolarsiccv":
            # Instantiate an object of LassoLarsIC() class
            self.sklearn_model = LassoLarsIC(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "linearregression" or \
                        self.r_sk == "linearregressioncv":
            # Instantiate an object of LinearRegression() class
            self.sklearn_model = LinearRegression(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "linearsvr" or \
                        self.r_sk == "linearsvrcv":
            # Instantiate an object of LinearSVR() class
            self.sklearn_model = LinearSVR(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "mlpregressor1" or \
                        self.r_sk == "mlpregressor1cv":
            # Instantiate an object of MLPRegressor() class (Scheme 1)
            self.sklearn_model = MLPRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "mlpregressor2" or \
                        self.r_sk == "mlpregressor2cv":
            # Instantiate an object of MLPRegressor() class (Scheme 2)
            self.sklearn_model = MLPRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "nusvr" or self.r_sk == "nusvrcv":
            # Instantiate an object of NuSVR() class
            self.sklearn_model = NuSVR(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "orthogonalmatchingpursuit" or \
            self.r_sk == "orthogonalmatchingpursuitcv":
            # Instantiate an object of OrthogonalMatchingPursuit() class
            self.sklearn_model = OrthogonalMatchingPursuit(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "passiveaggressiveregressor" or \
            self.r_sk == "passiveaggressiveregressorcv":
            # Instantiate an object of PassiveAggressiveRegressor() class
            self.sklearn_model = PassiveAggressiveRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "plscanonical" or \
            self.r_sk == "plscanonicalcv":
            # Instantiate an object of PLSCanonical() class
            self.sklearn_model = PLSCanonical(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "plsregression" or \
            self.r_sk == "plsregressioncv":
            # Instantiate an object of PLSRegression() class
            self.sklearn_model = PLSRegression(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "poissonregressor" or \
            self.r_sk == "poissonregressorcv":
            # Instantiate an object of PoissonRegressor() class
            self.sklearn_model = PoissonRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "quantileregressor" or \
            self.r_sk == "quantileregressorcv":
            # Instantiate an object of QuantileRegressor() class
            self.sklearn_model = QuantileRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "randomforestregressor" or \
                        self.r_sk == "randomforestregressorcv":
            # Instantiate an object of RandomForestRegressor() class
            self.sklearn_model = RandomForestRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "ransacregressor" or \
                        self.r_sk == "ransacregressorcv":
            # Instantiate an object of RANSACRegressor() class
            self.sklearn_model = RANSACRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "ridge" or self.r_sk == "ridgecv":
            # Instantiate an object of Ridge() class
            self.sklearn_model = Ridge(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "sgdregressor" or \
                        self.r_sk == "sgdregressorcv":
            # Instantiate an object of SGDRegressor() class
            self.sklearn_model = SGDRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "svr" or self.r_sk == "svrcv":
            # Instantiate an object of SVR() class
            self.sklearn_model = SVR(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "stackingregressor1" or \
                    self.r_sk == "stackingregressor1cv":
            # Instantiate an object of StackingRegressor() class
            estimators = [
                        ("lr", RidgeCV()),
                        ("svr", LinearSVR(random_state=1123581321))]
            self.sklearn_model = StackingRegressor(estimators=estimators,
                        final_estimator=RandomForestRegressor(n_estimators=10,
                        random_state=1123581321)
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "stackingregressor2" or \
                    self.r_sk == "stackingregressor2cv":
            # Instantiate an object of StackingRegressor() class
            estimators = [
                        (
                        "dtree",DecisionTreeRegressor(random_state=1123581321)),
                        ("xtree",ExtraTreeRegressor(random_state=1123581321)),
                        ("xtrees", ExtraTreeRegressor(random_state=1123581321))]
            self.sklearn_model = StackingRegressor(estimators=estimators,
                        final_estimator=RandomForestRegressor(n_estimators=10,
                        random_state=1123581321)
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "theilsenregressor" or \
                    self.r_sk == "theilsenregressorcv":
            # Instantiate an object of TheilSenRegressor() class
            self.sklearn_model = TheilSenRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "transformedtargetregressor1" or \
                    self.r_sk == "transformedtargetregressor1cv":
            # Instantiate an object of TransformedTargetRegressor() class
            self.sklearn_model = TransformedTargetRegressor(
                        regressor=LinearRegression(),
                        func=np.log, inverse_func=np.exp
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "transformedtargetregressor2" or \
                    self.r_sk == "transformedtargetregressor2cv":
            # Instantiate an object of TransformedTargetRegressor() class
            self.sklearn_model = TransformedTargetRegressor(
                        regressor=ExtraTreeRegressor(random_state=1123581321),
                        func=np.log, inverse_func=np.exp
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "tweedieregressor" or \
                    self.r_sk == "tweedieregressorcv":
            # Instantiate an object of TweedieRegressor() class
            self.sklearn_model = TweedieRegressor(**self.param_dict
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "votingregressor1" or \
                    self.r_sk == "votingregressor1cv":
            # Instantiate an object of VotingRegressor() class
            r1 = LinearRegression()
            r2 = RandomForestRegressor(n_estimators=10, random_state=1123581321)
            r3 = KNeighborsRegressor()
            self.sklearn_model = VotingRegressor([("lr",r1),("rf",r2),("r3",r3)]
                        ).fit(X_train,self.y_train)

        elif self.r_sk == "votingregressor2" or \
                    self.r_sk == "votingregressor2cv":
            # Instantiate an object of VotingRegressor() class
            r1 = DecisionTreeRegressor(random_state=1123581321)
            r2 = ExtraTreeRegressor(random_state=1123581321)
            r3 = ExtraTreeRegressor(random_state=1123581321)
            self.sklearn_model =VotingRegressor([("dt",r1),("xt",r2),("xts",r3)]
                        ).fit(X_train,self.y_train)

        # If cv call cross_validation() function
        if "cv" in self.r_sk:
            self.sklearn_model = cross_validation(self.sklearn_model,X_train,
                                        self.y_train,
                                        self.n_splits_cv,
                                        self.random_state_cv,
                                        False)

    # Define build() method
    def build(self):
        """Method to build multiple regression models"""

        # Import section
        import math
        import numpy as np
        from itertools import combinations
        from sklearn import preprocessing

        # Set up variables
        data_features = []
        n_models = 0
        y_list = [self.y_train,self.y_test]
        data_list = ["Training Set","Test Set"]
        data_out = self.header
        current_data = ""

        # Calculate the number of combinations (Cn,l)
        n = len(self.features)
        C_n_l = \
        int(math.factorial(n)/(math.factorial(self.l)*math.factorial(n-self.l)))
        rmse_array = np.zeros(C_n_l)

        # Get all combinations of elements in features and length l
        comb_features = combinations(self.features,self.l)

        # Show message
        msg_out = "\nGenerating (C"+str(n)+","+str(self.l)+" = "
        msg_out += str(C_n_l)+") regression models ("+self.r_sk+")"
        print(msg_out,end="...")

        # Looping through comb_features
        for current_features in list(comb_features):
            # Make it a list
            current_features = list(current_features)
            data_features.append(current_features)

            try:
                # For training set
                X_train = self.training_set[current_features]
                scaler_train = preprocessing.StandardScaler().fit(X_train)
                X_train = scaler_train.transform(X_train)

                # For test set
                X_test = self.test_set[current_features]
                scaler_test = preprocessing.StandardScaler().fit(X_test)
                X_test = scaler_test.transform(X_test)
                n_models += 1

                # Call pick_Regressor() method
                self.pick_Regressor(X_train)

                # Predictions
                y_train_pred_reg = self.sklearn_model.predict(X_train)
                y_test_pred_reg = self.sklearn_model.predict(X_test)
                y_pred_list = [y_train_pred_reg,y_test_pred_reg]

                # Looping through data
                for count,y in enumerate(y_list):
                    # Call Walsh_and_de_Azevedo_regression_metrics() function
                    r_pearson,p_pearson,rho,p_spearman,rss,rmse,mae,R2,dome,\
                    edomer2,edomerho,edome = \
                    Walsh_and_de_Azevedo_regression_metrics(y_list[count],
                                                            y_pred_list[count])

                    # Set up output line
                    line_out = self.r_sk+","+str(current_features).\
                            replace("'","").replace("[","").replace("]","").\
                            replace(","," ")
                    line_out += ","+data_list[count]+","+str(len(y))+","
                    line_out += "{:.4f}".format(r_pearson)+","
                    line_out += "{:.4e}".format(p_pearson)+","
                    line_out += "{:.4f}".format(r_pearson**2)+","
                    line_out += "{:.4f}".format(rho)+","
                    line_out += "{:.4e}".format(p_spearman)+","
                    line_out += "{:.4e}".format(rss)+","
                    line_out += "{:.4f}".format(rmse)+","+"{:.4f}".\
                                                                format(mae)+","
                    line_out += "{:.4f}".format(R2)+","+"{:.4f}".\
                                                                format(dome)+","
                    line_out += "{:.4f}".format(edomer2)+","
                    line_out += "{:.4f}".format(edomerho)+","
                    line_out += "{:.4f}".format(edome)

                    # Get test set
                    if count > 0:

                        # Add test set rmse to the array rmse_array
                        rmse_array[n_models-1] = rmse

                    # Update data_out and current_data
                    data_out += line_out+"\n"
                    current_data += line_out+"\n"

            except:
                msg_out = "\nError I didn't generate a regression model for "
                msg_out += str(current_features)
                print(msg_out)

        print("done!")

        # Open a new file and write data_out and close it
        c_out = str(len(self.features))+"_"+str(self.l)
        reg_out = self.r_sk+"_metrics_EDOME_C_"+c_out+".csv"
        fo_out = open(reg_out,"w")
        fo_out.write(data_out)
        fo_out.close()
        print("\nNumber of generated regression models: ",n_models)
        print("\nMetrics saved to file: "+reg_out)

        # Find minimum rmse
        min_i = np.argmin(rmse_array)
        best_features = data_features[min_i]
        best_rmse = rmse_array[min_i]
        msg_out = "\nMinimum RMSE: {:.4f}".format(best_rmse)
        msg_out += " for features: "+str(best_features)
        print(msg_out)

        # Regression model for the best set of features
        # For training set
        X_train = self.training_set[best_features]
        scaler_train = preprocessing.StandardScaler().fit(X_train)
        X_train = scaler_train.transform(X_train)

        # For test set
        X_test = self.test_set[best_features]
        scaler_test = preprocessing.StandardScaler().fit(X_test)
        X_test = scaler_test.transform(X_test)

        # Call pick_Regressor() method
        self.pick_Regressor(X_train)

        # Predictions
        y_train_pred_reg = self.sklearn_model.predict(X_train)
        y_test_pred_reg = self.sklearn_model.predict(X_test)

        # Call scatter_plot() function for the best regression model
        plt_out = self.r_sk+"_scatter_plot_test_set_C_"+c_out+".png"
        print("\nSaving scatter plot to file: "+plt_out,end="...")
        scatter_title = self.t_root
        scatter_title += " (RMSE: {:.4f}".format(best_rmse)+")"
        scatter_plot(self.y_test,y_test_pred_reg,
        self.n_points,self.point_type,
        self.color4data,self.color4models,
        self.scatter_xlabel,self.scatter_ylabel,scatter_title,
        self.xlabel_fontsize,self.ylabel_fontsize,self.title_fontsize,
        self.xticks_fontsize,self.yticks_fontsize,
        self.xlim,self.ylim,
        self.grid_on,
        plt_out,self.plt_dpi)
        print("done!")

        # Return current_data
        return current_data
```
